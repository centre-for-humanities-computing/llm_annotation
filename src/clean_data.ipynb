# Offensiveness

```python
off_3_eng = pd.read_csv("../Datasets_GPT_Output/Offensiveness/GPT3.5/EnglishOffensiveOutput.csv")
off_4_eng = pd.read_csv("../Datasets_GPT_Output/Offensiveness/GPT4/offensive-english-0-gpt-4-0-1.csv")
off_4t_eng = pd.read_csv("../Datasets_GPT_Output/Offensiveness/GPT4 Turbo/offensive-english-0-gpt-4-0125-preview-0-1.csv")
```

```python
off_3_tur = pd.read_csv("../Datasets_GPT_Output/Offensiveness/GPT3.5/offenseval-Turkish-GPT.csv")
off_4_tur = pd.read_csv("../Datasets_GPT_Output/Offensiveness/GPT4/offensive-turkish-0-gpt-4-0-1.csv")
off_4t_tur = pd.read_csv("../Datasets_GPT_Output/Offensiveness/GPT4 Turbo/offensive-turkish-0-gpt-4-0125-preview-0-1.csv",sep=";")
```

```python
def create_off_df(df1, df2, df3):
    full = df1.loc[: , "tweet":]
    full = full.rename({"gpt": "GPT3.5"},axis=1)

    full['GPT4'] = df2['offensive'].astype(int)
    full['GPT4-Turbo'] = df3['gpt4'].astype(int)

    return full
```

```python
off_3_eng = off_3_eng.rename({'text':"tweet"}, axis=1)

full_off_eng = create_off_df(off_3_eng, off_4_eng, off_4t_eng)
full_off_tur = create_off_df(off_3_tur, off_4_tur, off_4t_tur)
```

```python
full_off_eng.to_csv("../clean_data/offensive_twitter_english.csv", index=False)
full_off_tur.to_csv("../clean_data/offensive_twitter_turkish.csv", index=False)
```
