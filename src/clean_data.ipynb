## Discrete emotions

#### English

```python
eng_3_5 = pd.read_csv("../Datasets_GPT_Output/Discrete_Emotions/GPT3.5/emotion-class-english-3.5_temp=0.csv")
eng_4 = pd.read_csv("../Datasets_GPT_Output/Discrete_Emotions/GPT4/emotion-tweets-3-gpt-4-0-1.csv")
eng_4t = pd.read_csv("../Datasets_GPT_Output/Discrete_Emotions/GPT4 Turbo/emotion-english-0-gpt-4-0125-preview-0-1 (4).csv")
```

```python
eng_3_5.rename({"gpt": "GPT3.5"}, axis=1, inplace=True)
```

```python
label_dict_eng = {"anger": 1, "joy": 2, "sadness":3, "optimism":4, "love":4}


eng_4['GPT4'] = pd.from_dummies(eng_4[["anger", "joy", "sadness", "optimism"]])
eng_4 = eng_4.replace({"GPT4": label_dict_eng})
```

```python
eng_4t = eng_4t.rename({'gpt4':'gpt4t'}, axis = 1)
eng_4t
```

```python
full_eng = eng_3_5.loc[:, ["Tweet", "human", "GPT3.5"]]
full_eng['GPT4'] = eng_4['GPT4']
full_eng['GPT4_Turbo'] = eng_4t['gpt4t'].astype(int)
full_eng.head(10)
```

```python
full_eng.to_csv("../clean_data/emotion_twitter_english.csv", index=False)
```

#### Indonesian

```python
ind_3_5 = pd.read_csv("../Datasets_GPT_Output/Discrete_Emotions/GPT3.5/emotion-class-Indonesian-GPT-temp=0.csv")
ind_4 = pd.read_csv("../Datasets_GPT_Output/Discrete_Emotions/GPT4/emotion-indonesian-0-gpt-4-0-1.csv")
```

```python
ind_3_5.rename({"gpt": "GPT3.5"}, axis=1, inplace=True)

ind_3_5.head()
```

```python
label_dict_ind = {"anger": 1, "fear": 2, "sadness": 3, "love": 4, "joy": 5}


ind_4['GPT4'] = pd.from_dummies(ind_4[["anger", "fear", "sadness", "love", "joy"]])
ind_4 = ind_4.replace({"GPT4": label_dict_ind})
```

```python
full_ind = ind_4.loc[: , ["Tweet"]]
full_ind['human'] = ind_3_5['human']
full_ind['GPT3.5'] = ind_3_5['GPT3.5']
full_ind['GPT4'] = ind_4['GPT4']
```

```python
full_ind.to_csv("../clean_data/emotion_twitter_indonesian.csv", index=False)
```

# Moral foundations

```python
gpt4_auth = pd.read_csv("../Datasets_GPT_Output/Moral Foundations/GPT4/MFRC-authority (1)")
gpt4_care = pd.read_csv("../Datasets_GPT_Output/Moral Foundations/GPT4/MFRC-care")
gpt4_eq = pd.read_csv("../Datasets_GPT_Output/Moral Foundations/GPT4/MFRC-equality")
gpt4_loy = pd.read_csv("../Datasets_GPT_Output/Moral Foundations/GPT4/MFRC-loyalty")
gpt4_morsent = pd.read_csv("../Datasets_GPT_Output/Moral Foundations/GPT4/MFRC-moral_sentiment")
gpt4_prop = pd.read_csv("../Datasets_GPT_Output/Moral Foundations/GPT4/MFRC-proportionality")
gpt4_pur = pd.read_csv("../Datasets_GPT_Output/Moral Foundations/GPT4/MFRC-purity")
```

```python
gpt4_auth
```

```python
moral_full = gpt4_auth.loc[:, "text":]
```

```python
moral_full = moral_full.rename({"question_authority": "GPT4_Authority"}, axis=1)
moral_full['GPT4_Care'] = gpt4_care['questionCare']
moral_full['GPT4_Equality'] = gpt4_eq['question_equality']
moral_full['GPT4_Loyalty'] = gpt4_loy['question_loyalty']
moral_full['GPT4_MoralSentiment'] = gpt4_morsent['question_moral_sentiment']
moral_full['GPT4_Proportionality'] = gpt4_prop['question_proportionality']
moral_full['GPT4_Purity'] = gpt4_pur['question_purity']
```

```python
gpt4t_auth = pd.read_csv("../Datasets_GPT_Output/Moral Foundations/GPT4 Turbo/MFRC-authority-gpt-4-0125-preview")
gpt4t_eq = pd.read_csv("../Datasets_GPT_Output/Moral Foundations/GPT4 Turbo/MFRC-equality-gpt-4-0125-preview")
gpt4t_loy = pd.read_csv("../Datasets_GPT_Output/Moral Foundations/GPT4 Turbo/MFRC-loyalty-gpt-4-0125-preview")
gpt4t_morsent = pd.read_csv("../Datasets_GPT_Output/Moral Foundations/GPT4 Turbo/MFRC-moral_sentiment-gpt-4-0125-preview")
gpt4t_prop = pd.read_csv("../Datasets_GPT_Output/Moral Foundations/GPT4 Turbo/MFRC-proportionality-gpt-4-0125-preview")
gpt4t_pur = pd.read_csv("../Datasets_GPT_Output/Moral Foundations/GPT4 Turbo/MFRC-purity-gpt-4-0125-preview")
```

```python
moral_full['GPT4-Turbo_Authority'] = gpt4t_auth['question_authority']
moral_full['GPT4-Turbo_Equality'] = gpt4t_eq['question_equality']
moral_full['GPT4-Turbo_Loyalty'] = gpt4t_loy['question_loyalty']
moral_full['GPT4-Turbo_MoralSentiment'] = gpt4t_morsent['question_moral_sentiment']
moral_full['GPT4-Turbo_Proportionality'] = gpt4t_prop['question_proportionality']
moral_full['GPT4-Turbo_Purity'] = gpt4t_pur['question_purity']
```

```python
moral_full.to_csv("../clean_data/moral-foundations_reddit_english.csv", index=False)
```
